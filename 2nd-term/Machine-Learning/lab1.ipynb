## Simple linear regression

```python
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

diabetes = datasets.load_diabetes()
```

```python
diabetes.data.shape
```

```python
diabetes.target.shape
```

```python
X_train = diabetes.data[:400, 2:3]
y_train = diabetes.target[:400]
X_test = diabetes.data[400:, 2:3]
y_test = diabetes.target[400:]
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
print("coefficient: {}".format(model.coef_))
print("intercept: {}".format(model.intercept_))
```

```python
# plt.scatter(X_train, y_train, color='black')
plt.scatter(X_test, y_test, color='red')

plt.plot(X_test, model.predict(X_test), linewidth=3)
# x, y 축 단위 없애기
# plt.xticks(())
# plt.yticks(())

plt.show()
plt.close()
```

## Maximum Likelihood Estimation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib
```

```python
fig = plt.figure(figsize=(14, 8))
for i, N in enumerate([5, 10, 50, 500]):
    data = np.random.normal(loc=0, scale=1, size=N) # mean, std, N
    
    u = np.mean(data)
    s = np.sqrt(np.sum((data - u) ** 2) / N) # standard deviation
    
    orig = norm(0, 1)
    est = norm(u, s)
    
    subplot = fig.add_subplot(2, 2, i+1)
    subplot.set_title("N={}, u={:.2f}, s={:.2f}".format(N, u, s))

    line_x = np.arange(-5, 5.1, 0.1)
    subplot.plot(line_x, orig.pdf(line_x), color='black', linestyle='--')
    subplot.plot(line_x, est.pdf(line_x), color='red', linestyle='--')
    subplot.scatter(data, est.pdf(data), color='red')

plt.show()
plt.close()
```

## Boston housing data

```python
from sklearn import datasets, linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
boston = datasets.load_boston()
```

```python
boston.data.shape
```

```python
boston.target.shape
```

```python
DATA_RATIO = 0.7

X_train_len = int(boston.data.shape[0] * DATA_RATIO)
X_train = boston.data[:X_train_len, -1:]
y_train = boston.target[:X_train_len]
X_test = boston.data[X_train_len:, -1:]
y_test = boston.target[X_train_len:]
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# plt.scatter(X_train, y_train, color='black')
plt.scatter(X_test, y_test, color='red')

# plt.plot(X_train, model.predict(X_train), linewidth=3)
plt.plot(X_test, model.predict(X_test), linewidth=3)

plt.show()
plt.close()
```
